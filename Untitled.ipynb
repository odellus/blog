```python
import matplotlib.animation as animation

fig = plt.figure()

seq_len = 512
window_overlap = 64
chunk_size = 128
a = np.zeros((seq_len, seq_len))

ims = []
im = plt.imshow(a, animated=True)
ims.append([im])

def color_chunk(a, k, chunk_size, overlap):
    a[chunk_size*k - k*overlap:chunk_size*(k+1) - k*overlap,
      chunk_size*k - k*overlap:chunk_size*(k+1)- k*overlap] = 1.0
    return a

n_iters = seq_len // (chunk_size - window_overlap)
# Boolean variable to see if we need to do last chunk.
do_last_chunk = seq_len % (chunk_size - window_overlap) != 0

for k in range(n_iters):
    a = color_chunk(a, k, chunk_size, window_overlap)
    im = plt.imshow(a, animated=True)
    ims.append([im])

if do_last_chunk:
    a[-chunk_size:,-chunk_size:] = 1.0
    im = plt.imshow(a, animated=True)
    ims.append([im])
    
ani = animation.ArtistAnimation(fig, ims, interval=800, blit=True,
                                repeat_delay=10000)

ani.save('sliding_local_attention.gif')
```
